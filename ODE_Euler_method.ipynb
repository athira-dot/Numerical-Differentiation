Explicit Euler

```python
#explicit euler
import numpy as np
import matplotlib.pyplot as plt
h=0.0001
l=int(1/h)
t=np.linspace(0,2,l)
u=np.zeros(l)
v=np.zeros(l)
#initial condition
u[0]=1
v[0]=0
for i in range(1,l):
  u[i]=u[i-1]*(1+998*h)+1998*v[i-1]
  v[i]=v[i-1]*(1-1999*h)-999*u[i-1]
plt.plot(t,v,color="red")
#plt.plot(t,u)
plt.xlabel("t")
plt.ylabel("v")
plt.title(" v vs t -explicit euler")
plt.savefig("v vs t -explicit euler.png")
plt.show()
```

Implicit Euler

```python
#implicit euler
import numpy as np
import matplotlib.pyplot as plt
dtstab=0.001
h=dtstab/2
l=int(1/h)
t=np.linspace(0,0.1,l)
u=np.zeros(l)
v=np.zeros(l)
#z=np.zeros(l)
#initial condition
u[0]=1
v[0]=0
for i in range(1,l):
  p=1+(1998*h)*(999*h)/((1-h*998)*(1+h*1999))
  u[i]=(1/p)*((v[i-1]*1998*h/((1-h*998)*(1+h*1999)))+u[i-1]/(1-998*h))
  v[i]=(1/(1+1999*h))*(v[i-1]-999*h*u[i])
uo=2*np.exp(-t)-np.exp(-1000*t)
vo=-np.exp(-t)+np.exp(-1000*t)
plt.plot(t,v,color="red",label="v")
plt.plot(t,u,color="blue",label="u")
plt.plot(t,uo,color="green",label="u original")
plt.plot(t,vo,color="black",label="v original")
plt.legend()
plt.title("implicit euler")
plt.savefig("14implicit.png")
plt.show()
```
