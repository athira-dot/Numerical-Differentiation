# Implementation of RK4 METHOD

Harmonic oscillator

```python
import numpy as np
import matplotlib.pyplot as plt
t=0
h=0.01
l=int(50/h)
x=np.zeros(l)
v=np.zeros(l)
x[0]=1
v[0]=0
t=np.linspace(0,50,l)
for i in range(1,l):
  k1=h*v[i-1]
  k2=h*v[i-1]
  k3=h*v[i-1]
  k4=h*v[i-1]
  x[i]=x[i-1]+(1/6)*(k1+2*k2+2*k3+k4)
  k1=-h*x[i-1]
  k2=-h*x[i-1]
  k3=-h*x[i-1]
  k4=-h*x[i-1]
  v[i]=v[i-1]+(1/6)*(k1+2*k2+2*k3+k4)
plt.plot(t,x)
plt.xlabel("t")
plt.ylabel("x")
plt.title("harmonic oscillator")
plt.savefig("foura.png")
plt.show()
```

Anarmonic oscillator

```python
t=0
h=0.01
l=int(50/h)
x=np.zeros(l)
v=np.zeros(l)
x[0]=1.5
v[0]=0
t=np.linspace(0,50,l)
for i in range(1,l):
  
  k1=-h*(x[i-1]**3)
  k2=-h*(x[i-1]**3)
  k3=-h*(x[i-1]**3)
  k4=-h*(x[i-1]**3)
  v[i]=v[i-1]+(1/6)*(k1+2*k2+2*k3+k4)
  k1=h*v[i-1]
  k2=h*v[i-1]
  k3=h*v[i-1]
  k4=h*v[i-1]
  x[i]=x[i-1]+(1/6)*(k1+2*k2+2*k3+k4)
plt.plot(t,x)
plt.xlabel("t")
plt.ylabel("x")
plt.title("anharmonic oscillator")
plt.savefig("fourc.png")
plt.show()
```
