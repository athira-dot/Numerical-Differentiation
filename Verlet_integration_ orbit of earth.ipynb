# verlet integration to find the orbit of earth

```python
import numpy as np
import matplotlib.pyplot as plt
M=1.989*10**30
G=6.67408*10**(-11)
h=3600 #should set
l=int(365*24*3600/h)
t=np.linspace(0,100*24*3600,l)
x=np.zeros(l)
y=np.zeros(l)
vx=np.zeros(l)
vy=np.zeros(l)
#initial conditions
x[0]=0       
y[0]=1.4710*10**11
vx[0]=3.0287*10**(4)
vy[0]=0
v1=vx[0]
v2=vy[0]
# verlet integration
for i in range(1,l):
  v1=v1+((h/2)*-G*M*x[i-1]/((np.sqrt(x[i-1]**2+y[i-1]**2))**3))
  x[i]=x[i-1]+h*(v1)
  v2=v2+(h/2)*-G*M*y[i-1]/((np.sqrt(x[i-1]**2+y[i-1]**2))**3)
  y[i]=y[i-1]+h*(v2)

  vx[i]=v1+(h/2)*-G*M*x[i]/((np.sqrt(x[i]**2+y[i]**2))**3)
  v1=vx[i]
  vy[i]=v2+(h/2)*-G*M*y[i]/((np.sqrt(x[i]**2+y[i]**2))**3)
  v2=vy[i]
plt.plot(x,y)
plt.xlabel("x")
plt.ylabel("y")
plt.title("Orbit of the Earth")
plt.savefig("tena.png")
plt.show()
```
